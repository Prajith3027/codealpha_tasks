```python
import pandas as pd

df = pd.read_csv('../data/netflix_titles.csv')

df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.shape
```

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows')
plt.show()
```

```python
df['rating'].value_counts()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(
    df['date_added'].str.strip(),
    errors='coerce'
)
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['year_added'].value_counts().sort_index()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

### Observation

The genre analysis reveals that Netflix offers a diverse range of content across multiple categories. The most common genres appear more frequently than others, indicating the platform's focus on providing popular and widely watched content. This distribution suggests that Netflix caters to a broad audience with varying entertainment preferences.

## Key Insights

1. Movies significantly outnumber TV Shows on Netflix.
2. TV-MA is one of the most common content ratings.
3. The United States contributes the highest amount of Netflix content.
4. The number of titles added to Netflix increased rapidly after 2015.
5. Popular genres such as International Movies, Dramas, and Comedies dominate the platform's catalog.

## Conclusion

This exploratory data analysis of the Netflix dataset provided insights into content distribution, ratings, countries of origin, release trends, and genres. The analysis shows that Netflix maintains a diverse content library with a strong emphasis on movies and internationally distributed content. The platform has experienced substantial growth in content additions over recent years, reflecting its expansion and global reach.
