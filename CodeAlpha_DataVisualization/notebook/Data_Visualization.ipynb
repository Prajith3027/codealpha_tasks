## Visualization 1: Sales by Category

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Category',
    y='Sales',
    data=df,
    estimator=sum
)

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```

### Observation

Technology generated the highest sales among all product categories, indicating strong customer demand. Furniture and Office Supplies also contributed significantly to overall sales.

## Visualization 2: Profit by Category

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Category',
    y='Profit',
    data=df,
    estimator=sum
)

plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.show()
```

### Observation

Technology generated the highest profit among all categories, while Furniture contributed comparatively lower profits despite having substantial sales.

## Visualization 3: Sales by Region

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Region',
    y='Sales',
    data=df,
    estimator=sum
)

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

### Observation

Sales performance varies across regions. Some regions contribute significantly more sales than others, indicating differences in customer demand and market potential.

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Region',
    y='Sales',
    data=df,
    estimator=sum
)

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.savefig('../images/sales_by_region.png', bbox_inches='tight')

plt.show()
```

## Visualization 4: Monthly Sales Trend

This visualization analyzes monthly sales performance and helps identify seasonal trends and fluctuations in revenue throughout the year.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

### Observation

Sales fluctuate throughout the year, with certain months generating higher revenue than others. This indicates seasonal variations in customer purchasing behavior.

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.savefig('../images/monthly_sales_trend.png', bbox_inches='tight')

plt.show()
```

## Visualization 5: Top 10 Sub-Categories by Sales

This visualization identifies the top-performing sub-categories based on total sales revenue.

```python
top_sub = (
    df.groupby('Sub-Category')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_sub
```

```python
plt.figure(figsize=(10,5))

top_sub.plot(kind='bar')

plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')

plt.show()
```

### Observation

A small number of sub-categories contribute a significant portion of total sales. These products represent the key revenue drivers of the business.

```python
plt.figure(figsize=(10,5))

top_sub.plot(kind='bar')

plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')

plt.savefig('../images/top10_subcategories_sales.png', bbox_inches='tight')

plt.show()
```

## Visualization 6: Sales vs Profit

This scatter plot shows the relationship between sales and profit for individual transactions.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='Sales',
    y='Profit',
    data=df
)

plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.show()
```

### Observation

Higher sales do not always result in higher profits. Some transactions generate high sales but low profits, indicating differences in profit margins and discount strategies.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='Sales',
    y='Profit',
    data=df
)

plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.savefig('../images/sales_vs_profit.png', bbox_inches='tight')

plt.show()
```

## Visualization 7: Correlation Heatmap

This heatmap displays the relationships among key numerical variables in the dataset.

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

### Observation

Discount shows a negative relationship with profit, suggesting that larger discounts can reduce profitability. Sales and profit exhibit a positive but moderate relationship.

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.savefig('../images/correlation_heatmap.png', bbox_inches='tight')

plt.show()
```

## Visualization 8: Profit Distribution

This histogram illustrates the distribution of profit values across all transactions.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Profit'],
    bins=30,
    kde=True
)

plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.show()
```

### Observation

Most transactions generate small profits, while a smaller number of transactions result in very high profits or losses. This indicates an uneven profit distribution across orders.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Profit'],
    bins=30,
    kde=True
)

plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.savefig('../images/profit_distribution.png', bbox_inches='tight')

plt.show()
```
